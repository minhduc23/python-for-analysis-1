## Drop Rows with Null Values

```python
nba = pd.read_csv('nba.csv')
nba.head(3)
```

```python
nba.tail(3)
```

```python
# Remove rows have the null value
nba.dropna()
```

### The "how" parameter is set to defaul at any, which means it will remove any rows that have null value. If you set to "all", it will only remove rows that contain all null values

```python
nba.dropna(how = 'all', inplace = True)

# Only the last row is removed because it has all null values.
```

```python
nba
```

```python
# Remove any columns that have the null values
nba.dropna(axis = 1) 

# instead of using 1, you can use "columns"
nba.dropna(axis = 'columns')
```

```python
# It will only drop rows where the salary is null
nba.dropna(subset = ['Salary'])
```

## The .astype() Method

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba.tail(3)
```

```python
nba['Salary'].fillna(0, inplace = True)
nba['College'].fillna("None", inplace = True)
nba.head(6)
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
nba['Salary'] = nba['Salary'].astype("int")
```

```python
nba.info()
```

```python
nba['Number'] = nba['Number'].astype("int")
nba['Age'] = nba['Age'].astype('int')
```

```python
nba.info()
```

```python
nba['Position'].nunique()
```

```python
nba['Position'] = nba['Position'].astype('category')
```

```python
nba.head(10)
nba.info()
```

## Sort a DataFrame with the .sort_values() method: PART I

```python
nba = pd.read_csv('nba.csv')
```

```python
nba.sort_values(by = 'Name')
nba.sort_values(by = 'Age', ascending = False)
nba.sort_values(by = 'Salary', ascending = False, inplace = True)
nba.head(3)
```

```python
nba.sort_values(by = 'Salary', na_position = "first")
```

## Sort a DataFrame with the .sort_values() method: PART II

```python
nba = pd.read_csv('nba.csv')
```

```python
nba.sort_values(by = ['Team', 'Name'], ascending = [True, False])
```

## Rank values with the .rank() method

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.tail(3)
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending = False).astype('int')
nba.head()
```

```python
nba.sort_values(by = 'Salary', ascending = False)
```
